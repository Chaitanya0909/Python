```python
#droping columns with NAs(beacuse there are so many)
data_drop = data.drop(data[["NumberDosesReceived", "FirstDoseRefused", "Denominator"]], axis = 1)
data_drop
```

```python
#taking count,mean,min,max of data_drop and then checking the population
a = pd.DataFrame(data_drop.groupby("ReportingCountry")["Population"].aggregate(["mean", "count","min", "max"]))
a
```

```python
# introducing a new column as sum of first + second + unknown dose

data_drop.insert(4, "Total Dose",(data_drop["FirstDose"] + data_drop["SecondDose"]+data_drop["UnknownDose"]))
data_drop
```

```python
# Taking total sum of firstdose (fds), seconddose (sds), unknowndose (uds), total dose(tds)
```

```python
fds = sum(data_drop["FirstDose"])
fds
```

```python
sds = sum(data_drop["SecondDose"])
sds
```

```python
uds = sum(data_drop["UnknownDose"])
uds
```

```python
tds = sum(data_drop["Total Dose"])
tds
```

```python
#creating a pie chart

ml = ["FirstDose", "SecondDose", "UnknownDose"]
plt.pie((fds,sds,uds), labels = ml, autopct = "%.2f%%",explode = [.1,.1,.1],shadow = True, radius=1);

plt.xlabel("Total number of doses across the world")

plt.legend(
    loc='best',
    labels= ml);
```
